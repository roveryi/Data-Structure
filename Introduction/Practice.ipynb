# Maximum Subsequence
Given an array, return the Subsequence with the maximum sum


```python
# Method 1: Brute Force
# Complexity: O(n^3)
def MaxSubseqSum1(array):
        n = len(array)
        ans = 0
        for i in range(n):
                for j in range(i+1,n):
                        ans_cur = 0
                        for k in range(i,j+1):
                                ans_cur += array[k]
                        ans = max(ans_cur, ans)
        return ans 


# Method 2: Brute Force
# Complexity: O(n^2)
def MaxSubseqSum2(array):
        n = len(array)
        ans = 0

        for i in range(n):
                for j in range(i+1,n):
                        ans = max(ans, sum(array[i:j+1]))
        return ans

# Method 3: Divide and Conqure 
# Complexity: O(nlogn)
# The optimal solution is given by the maximum of the left array, right array and an max array that cross midpoint 
def MaxSubseqSum3(array, left, right):
        if left == right:
                return max(array[left], 0)
        if left == right - 1:
                return max(array[left], array[right], sum(array[left:right+1]), 0)

        mid = int((left + right)/2)
        MaxLeft = MaxSubseqSum3(array, left, mid)
        MaxRight = MaxSubseqSum3(array, mid, right)

        MaxLeftBorder, MaxRightBorder = 0, 0
        LeftBorder, RightBorder = 0, 0 

        i = mid
        while i >= left:
                LeftBorder += array[i]
                MaxLeftBorder = max(LeftBorder, MaxLeftBorder)
                i -= 1
        i = mid+1
        while i <= right:
                RightBorder += array[i]
                MaxRightBorder = max(RightBorder, MaxRightBorder)
                i += 1
        
        return max(MaxLeft, MaxRight, MaxLeftBorder+MaxRightBorder)

# Method 4: Greedy
# Complexity: O(n)
# Concept: if the sum of current and previsou values is positive, then it 'contributes' to the increasing of the total sum, then it can be added to the next one. 
def MaxSubseqSum4(array):
        n = len(array)
        for i in range(1,n):
                array[i] = max(array[i-1]+array[i], 0)
        return max(array)
```

# Binary Search
Given sorted array, search target in O(logn)

```python
def BinarySearch(nums, target):
    left, right, mid = 0, len(nums)-1, int(len(nums)/2)

    while left < right:
        # Return part
        if target == nums[left]:
            return left
        if target == nums[right]:
            return right 
        if target == nums[mid]: 
            return mid 
        # Recursion part
        if target < nums[mid]:
            right = mid-1
            mid = int((left+right)/2)

        elif target > nums[mid]:
            left = mid+1
            mid = int((left+right)/2)

    return None 
```
